```python
def ignore(img):
    largura, altura, dimensoes = img.shape
    img = img.astype(np.uint8)
    for d in range(dimensoes): # check each channel:
        for l in range(largura): # check each line:
            for a in range(altura): # check each pixel:
                if l != 0 and l != largura-1 and a != 0 and a != altura-1:
                    img[l][a][d] = 255
    
    return img
```

```python
img2 = ignore(img.imgRed)
```

```python
img.manyShowWithTitle({'img': img.imgRed, 'title': 'Original'}, {'img': img2, 'title': 'Ignorada'})
```
